```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4, n_features=1, n_informative=1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# Apllying OLS (Ordinary latest Squared)
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red')
```

```python
# Lets apply Gradient Decent assuming slope is contant m = 78.35
# and let's assume the starting value for intercept b = 0
```

```python
y_pred = ((78.35 * X) + 0)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='blue',label='b = 0')
plt.legend()
plt.show()
```

```python
m = 78.35
b = 0
# Iteration 1
Loss_slope = -2 * np.sum(y - m*X.ravel() - b)
Loss_slope
```

```python
# Let's take a learning rate = 0.1
lr = 0.1

step_size = Loss_slope*lr
step_size
```

```python
# Calculating the new intercept
b = b - step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1, color='green', label='b = {}'.format(b))
plt.plot(X,y_pred,color='blue',label='b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 2
Loss_slope = -2 * np.sum(y - m*X.ravel() - b)
Loss_slope
```

```python
step_size = Loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1, color='green', label='b = {}'.format(b))
plt.plot(X,y_pred,color='blue',label='b = 0')
plt.legend()
plt.show()
```
